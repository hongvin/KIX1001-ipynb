```python
import sympy as sy
from sympy.abc import x,y,t,r,h                   # variable x
from sympy.solvers import solve                   # Solving equations
from sympy import sqrt,tan,sin,cos,sec,pi,root,ln
from sympy import log,exp,atan,asinh,atanh,asin   # Import all required math function
from sympy import diff, idiff                     # Solving differentiation
sy.init_printing(use_latex=True)                  # Show it in natural display
```

### Question 1(a)
$$\displaystyle \lim _{x \rightarrow 5} \frac{x^{2}-25}{x^{2}+x-30}$$

```python
lim = 5
func = (x**2-25)/(x**2+x-30)
sy.limit(func,x,lim)
```

### Question 2
Find the derivative of $\log(4+\cos x)$.

### Question 3
Find $\frac{dy}{dx}$ for $\cos(x^2)=xe^y$.

```python
eqn=cos(x**2)-x*exp(y)
idiff(eqn,y,x)
```

### Question 4
Find $\frac{dy}{dx}$ for $x^3y^3-2y=x$.

```python
eqn=x**3*y**3-2*y-x
idiff(eqn,y,x)
```

### Question 5
A curve in the plane is defined parametrically by the equations $x=7\ln(t)$ and $y=\sqrt{1-4t}$. Find $\frac{dy}{dx}$.

```python
dydt = sqrt(1-4*t).diff(t)
dxdt = 7*ln(t).diff(t)
dydt/dxdt
```

### Question 6
A curve in the plane is defined parametrically by the equations $x=t^2-1$ and $y=2e^t$. Find $\frac{dy}{dt}$.

```python
dydt = 2*exp(t).diff(t)
dxdt = (t**2-1).diff(t)
dydt/dxdt
```

### Question 7(a)
Find $y'$ for $x^2\tan(y)+y^{10}\sec(x)=2x$

```python
eqn=x**2*tan(y)+y**10*sec(x)-2*x
idiff(eqn,y,x)
```

### Question 7(b)
Find $y'$ for $x^3y^5+3x=8y^3+1$

```python
eqn=x**3*y**5+3*x-8*y**3-1
idiff(eqn,y,x)
```

### Question 7(c)
Find $y'$ for $e^{2x+3y}=x^2-\ln(xy^3)$

```python
eqn=exp(2*x+3*y)-x**2+ln(x*y**3)
idiff(eqn,y,x)
```

### Question 8

### Question 9
Find $y'$ for $10e^{2xy}=e^{15y}+e^{13x}$.

```python
eqn=10*exp(2*x*y)-exp(15*y)-exp(13*x)
idiff(eqn,y,x)
```
